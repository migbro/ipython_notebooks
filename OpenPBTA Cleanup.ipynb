```python
import sevenbridges as sbg
from sevenbridges.errors import SbgError
from sevenbridges.http.error_handlers import rate_limit_sleeper, maintenance_sleeper
import sys
import re
import pdb
import concurrent.futures
from requests import request
import datetime
config = sbg.Config(profile='turbo')
api = sbg.Api(config=config, error_handlers=[rate_limit_sleeper, maintenance_sleeper])
```

## Patch metadata
The tasks have 0 metadata, so patching to make life easier

```python
def add_meta_to_outputs(task):
    try:
        if in_key in task.inputs and re.search(phrase, task.name):
            metadata = {}
            if tag_flag:
                tags = tag_list
            in_file = api.files.get(task.inputs[in_key])
            for key in in_file.metadata:
                metadata[key] = in_file.metadata[key]
            # outputs for these tasks are file arrays
            for key in task.outputs:
                if isinstance(task.outputs[key], list):
                    for out_file in task.outputs[key]:
                        try:
                            out_file_obj = api.files.get(out_file.id)
                        except Exception as e:
                            print (e)
                            print ("Error getting file output for " + key + " " + task.name + " " + task.id + " skipping!")
                            continue
                        if out_file_obj.metadata is None or len(out_file_obj.metadata) == 0:
                            out_file_obj.metadata = metadata
                            if tag_flag:
                                out_file_obj.tags = tags
                            out_file_obj.save()
                else:
                    try:
                        out_file_obj = api.files.get(task.outputs[key].id)
                        if out_file_obj.metadata is None or len(out_file_obj.metadata) == 0:
                            out_file_obj.metadata = metadata
                            if tag_flag:
                                out_file_obj.tags = tags
                            out_file_obj.save()
                    except Exception as e:
                        print(e)
                    
            return 0
        else:
            return None
    except Exception as e:
        print (e)
        print ("Got an error processing " + task.name + " " + task.id)
        exit(1)
```

```python
project = 'cavatica/qtx6-ewhc'
tasks = api.tasks.query(project=project, status="COMPLETED").all()
i = 1
n = 100
phrase = 'snpEff_vcf_2_Annovar'
in_key = 'input_vcf'
tag_flag = False
with concurrent.futures.ThreadPoolExecutor(16) as executor:
    results = {executor.submit(add_meta_to_outputs, task): task for task in tasks}
    for result in concurrent.futures.as_completed(results):
        if i % n == 0:
            sys.stderr.write(str(i) + ' tasks processed\n')
        if result.result() is not None:
            i += 1
# for task in tasks:
#     add_meta_to_outputs(task)
```

## Run clinvar annot + pathogenicity filtering

```python
def setup_refs():
    ref_dict = {}
    ref_dict['db_file'] = api.files.get('62e969134977726e6fd50092')
    ref_dict['db_name'] = 'clinvar'
    ref_dict['fields'] = 'CLNSIG,CLNREVSTAT'
    ref_dict['mode'] = 'annotate'
    ref_dict['include_expression'] = '(INFO/CLNSIG="Pathogenic" | INFO/CLNSIG="Likely_pathogenic") && (INFO/CLNREVSTAT="practice_guideline"| INFO/CLNREVSTAT="reviewed_by_expert_panel"| INFO/CLNREVSTAT="criteria_provided,_multiple_submitters,_no_conflicts") && INFO/gnomAD_exome_ALL < 0.001 && INFO/gnomAD_exome_AFR < 0.001 && INFO/gnomAD_exome_AMR < 0.001 && INFO/gnomAD_exome_ASJ < 0.001 && INFO/gnomAD_exome_EAS < 0.001 && INFO/gnomAD_exome_FIN < 0.001 && INFO/gnomAD_exome_NFE < 0.001 && INFO/gnomAD_exome_OTH < 0.001 && INFO/gnomAD_exome_SAS < 0.001 && INFO/gnomAD_genome_ALL < 0.001 && INFO/gnomAD_genome_AFR < 0.001 && INFO/gnomAD_genome_AMR < 0.001 && INFO/gnomAD_genome_ASJ < 0.001 && INFO/gnomAD_genome_EAS < 0.001 && INFO/gnomAD_genome_FIN < 0.001 && INFO/gnomAD_genome_NFE < 0.001 && INFO/gnomAD_genome_OTH < 0.001 && (INFO/ANN~"|EXO1|" | INFO/ANN~"|LIG1|" | INFO/ANN~"|MLH1|" | INFO/ANN~"|MLH3|" | INFO/ANN~"|MSH2|" | INFO/ANN~"|MSH3|" | INFO/ANN~"|MSH6|" | INFO/ANN~"|PCNA|" | INFO/ANN~"|PMS2|" | INFO/ANN~"|POLD1|" | INFO/ANN~"|POLD2|" | INFO/ANN~"|POLD3|" | INFO/ANN~"|POLD4|" | INFO/ANN~"|RFC1|" | INFO/ANN~"|RFC2|" | INFO/ANN~"|RFC3|" | INFO/ANN~"|RFC4|" | INFO/ANN~"|RFC5|" | INFO/ANN~"|RPA1|" | INFO/ANN~"|RPA2|" | INFO/ANN~"|RPA3|" | INFO/ANN~"|RPA4|" | INFO/ANN~"|SSBP1|" | INFO/ANN~"|TP53|")'
    return ref_dict
```

```python
project = 'd3b-bixu/dev-wgsa'
app = project + '/clinvar-pathogenic-filter'
inputs = api.files.query(project=project, tags = ['annovar'])
refs = setup_refs()
for file_input in inputs:
    task_name = 'PATHOGENICITY PIPELINE: ' + file_input.metadata['Kids First Biospecimen ID']
    in_dict = {}
    for key in refs:
        in_dict[key] = refs[key]
    in_dict['input_vcf'] = file_input
    task = api.tasks.create(
            name=task_name, project=project, app=app, inputs=in_dict,
            run=False
        )
    task.inputs['output_basename'] = task.id
    task.save()
```
